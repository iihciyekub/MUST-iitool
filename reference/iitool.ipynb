# 对生成的 bbl 文件进行

```python
# 增加英文文献分隔字段到bbl文件中
c1 = r"\bibitem[En, 2019]{en2019}{\fontsize{16pt}{\baselineskip}\selectfont{\it\bfseries 英文文獻}}"
# 增加中文文献分隔字段到bbl 文件中
c2 = r"\bibitem[zh, 2019]{zh2019}{\fontsize{16pt}{\baselineskip}\selectfont{\it\bfseries 中文文獻}}"
# 判斷是否存在中文字符
zhmodel = re.compile(u'[\u4e00-\u9fa5]')
```

```python
# 搜索本地目錄 找到 bbl 文件
bblfile = findfile(suffix='bbl')
bbltext = ""
# 讀取 bbl 文件
with open(bblfile, "r", encoding='utf-8') as f:
    bbltext = f.read()

# 對 bbl 文件裏的文本進行分割
bbltextlist = bbltext.split(r"\bibitem[")
bbltextlist = bbltextlist[1:]
bbltextlist[-1] = bbltextlist[-1].replace(r"\end{thebibliography}", "")
```

```python
# 利 findindx() 方法在 dic 字典中索引出作者名的序号-->创建对应的新字典 d;
d = {}
for i in bbltextlist:
    d[i] = findindx(dic, i[0])

# 对新字典 d 按字典值进行升序 排序~
tou = sorted(d.items(), key=lambda d: d[1], reverse=False)
```

```python
# 定义一个 re 模式,用于判断是否存在中文字符
newbbltex = "\\begin{thebibliography}{}\n\n\n"+f"{c1}\n\n\n"
addc = True
for i, k in tou:
    # 判斷是否包含中文
    match = zhmodel.search(i)

    # 如果 包含中文則替換到一些英文為中文
    if match:
        i = i.replace("et~al.", "等")
        i = i.replace("and ", "與")
    i = r"\bibitem["+i

    # 是否第一次出現中文
    if match and addc:
        newbbltex = newbbltex + "\n" + f"{c2}\n\n\n"
        addc = False
    newbbltex += i
newbbltex += "\n\\end{thebibliography}"

# 更新本地 bbl 文件
with open(bblfile, "w", encoding='utf-8') as f:
    f.write(newbbltex)
```

```python
# 对 bib 文件进行逐行读取,并将内容格式化
def allbibFileToStr(bibfilePath):
    '''
    方法:将所有 bib 文件的内容进行格式化并读取出来,返回文本
    bibfilePath: 包含 bib 文件的路径.str & list
    '''
    bibFileStr = ""
    # 如果 bibfilePath 參數為 str 時,則處理為 list 類型
    if type(bibfilePath) == type(""):
        bibfilePath = [bibfilePath]

    # 讀取所有 bib 文件的內容
    for refF in bibfilePath:
        with open(refF, 'r', encoding="UTF-8") as f:
            while 1:
                # 逐行读取
                a = f.readline()

                # 去除空行
                if a == '\n':
                    continue

                # 去除两边空格
                temp = a.strip()

                if temp == None:
                    continue
                # 如果該行包含 = ,就在前面加個 tab 個縮進
                if "=" in temp:
                    temp = "\t" + temp

                # 处理最后一条数据包含逗号的情况
                if temp == "}":
                    if bibFileStr[-2] == ",":
                        bibFileStr = bibFileStr[:-2]+"\n"
                bibFileStr += (temp+"\n")
                if not a:
                    break

        # 处理@字段与新行起写的问题
        bibFileStr = bibFileStr.replace("}@", "}\n@")
    # 返回所有 bib 文件的文本內容
    return bibFileStr


# 传入字符计算其对应的参考文献排序的数值
def sortbibItem(tag):
    '''
    返回指定字符在字典 dic 中的 MUST 参考文献排序数值
    '''
    global dic
    if tag == "":
        return 0
    else:
        return list(dic.keys())[list(dic.values()).index(tag)]


def bibItemisZH(tag):
    """
    判斷 bibitem 是否存在中文
    """
    zhmodel = re.compile(u'[\u4e00-\u9fa5]')
    match = zhmodel.search(tag)
    return 'chinese' if match else ""


# 将bibitem的文本内容进行解析,并存入bibDict 字典中
def bibItemTobibDict(bibitemStr):
    '''
    对 bibitem 的文本内容进行解析,并增加到字典中
    '''
    global bibDict
    # 提取 bibitem中的 引用标签字符串
    x = bibitemStr.split(",")[0].split("{")
    # 如果引用标签已经存在刚不
    if x[1].strip() not in bibDict["citelabel"]:
        bibDict["bibType"].append(x[0].strip().lower())
        bibDict["citelabel"].append(x[1].strip())
        # 基本屬性添加
        for i in bibAttribute:
            if i in ["bibType", "citelabel"]:
                continue
            # 匹配屬性值 並添加到字典中
            x = re.findall(i+".*?{(.*?)}", bibitemStr, re.I)
            bibDict[i].append("" if x == [] else x[0])
```

```python
def bibPDtobblFileText_ZH(i, bibPD):
    # 對 pd 格式的數據進行處理
    if "and" in bibPD.author[i]:
        authorlist = [si.strip() for si in bibPD.author[i].split(" and ")]
    elif "," in bibPD.author[i]:
        authorlist = [si.strip() for si in bibPD.author[i].split(" and ")]
    else:
        authorlist = [bibPD.author[i]]
    # 出現在正文中的引用標籤作者名
    citeauthor = authorlist[0]
    if len(authorlist) >= 3:
        citeauthor += "等人"
    elif len(authorlist) == 2:
        citeauthor += "與"
        citeauthor += authorlist[1]
    # 出現在參考文獻中的作者信息
    refauthor = authorlist[0]
    if len(authorlist) >= 2:
        if len(authorlist) >= 8:
            for j in authorlist[1:5]:
                refauthor += "、"
                refauthor += j
            refauthor += "、…"
        elif len(authorlist) >= 3:
            for j in authorlist[1:-1]:
                refauthor += "、"
                refauthor += j
        refauthor += "與"
        refauthor += authorlist[-1]

    #　----------------------------------------->>
    ToutputC = ""
    # type I  ------->>普通文章类
    if bibPD.bibType[i] in ["article", "misc", "manual"]:
        # 格式化文本
        ToutputC += (r"\bibitem[" + citeauthor + ", ")
        ToutputC += (bibPD.year[i] + r"]{")
        ToutputC += (bibPD.citelabel[i] + "}\n")
        ToutputC += (refauthor + r"（" + bibPD.year[i] + "）。\n")
        ToutputC += (r"\newblock " + bibPD.title[i])
        #  判斷是否存在期刊名
        if bibPD.journal[i] != "":
            ToutputC += "。\n"
            ToutputC += r"\newblock \textbf{" + bibPD.journal[i] + r"}"
        if bibPD.number[i] != "" or bibPD.volume[i] != "":
            ToutputC += "，"
            if bibPD.volume[i] != "":
                ToutputC += (r"\textbf{" + bibPD.volume[i] + r"}")
            # 判斷是否存卷,卷號
            if bibPD.number[i] != "":
                ToutputC += (r"\textbf{(" + bibPD.number[i] + r")}")
        # 判斷是否存在頁碼
        if bibPD.pages[i] != "":
            ToutputC += f"，{bibPD.pages[i]}"
        ToutputC += "。"
        if bibPD.doi[i] != "":
            ToutputC = ToutputC + f" doi:{bibPD.doi[i]}"
        ToutputC += "\n\n"

    # type II  ------->>畢業論文
    elif bibPD.bibType[i] in ["thesis", "phdthesis", "mastersthesis"]:
        ToutputC += (r"\bibitem[" + citeauthor + ", ")
        ToutputC += (bibPD.year[i] + r"]{")
        ToutputC += (bibPD.citelabel[i] + "}\n")
        ToutputC += (refauthor + r" (" + bibPD.year[i] + ")。\n")
        ToutputC += r"\newblock \textbf{"
        ToutputC += bibPD.title[i] + "}。"

        if bibPD.bibType[i] == "phdthesis":
            ToutputC += "博士論文"
        elif bibPD.bibType[i] == "mastersthesis":
            ToutputC += "碩士論文"
        elif bibPD.bibType[i] == "thesis":
            ToutputC += "論文"
        # 判断是否存在学校信息
        if bibPD.school[i] != "":
            ToutputC += ("，" + bibPD.school[i])
        elif bibPD.university[i] != "":
            ToutputC += ("，" + bibPD.university[i])
        ToutputC += "。\n\n"

    # type II  ------->>online
    elif bibPD.bibType[i] in ["online"]:
        if bibPD.date[i] != "":
            date = bibPD.date[i].split("-")
            if len(date) == 3:
                date = f"{date[0]}年{date[1]}月{date[2]}日"
            elif len(date) == 2:
                date = f"{date[0]}年{date[1]}月"
            elif len(date) == 1:
                date = f"{date[0]}年"
        else:
            date = "无日期"
        ToutputC += (r"\bibitem[" + bibPD.citelabel[i] + ", ")
        ToutputC += (date + r"]{")
        ToutputC += (bibPD.citelabel[i] + "}\n")
        if bibPD.author[i] == "":
            ToutputC += (bibPD.title[i]+"。(")
            ToutputC += (date + ")。取自 ")
            ToutputC += bibPD.url[i]
        else:
            ToutputC += (bibPD.author[i]+"(")
            ToutputC += (date + ")【" +
                         bibPD.description[i] + "】。" + bibPD.title[i])
            ToutputC += ("。取自 " + bibPD.url[i])
        ToutputC += "\n\n"

    # type ------->> 会议论文
    elif bibPD.bibType[i] in ["proceedings", "inproceedings", "conference"]:
        ToutputC += (r"\bibitem[" + citeauthor + ", ")
        ToutputC += (bibPD.year[i] + r"]{")
        ToutputC += (bibPD.citelabel[i] + "}\n")
        ToutputC += (refauthor + r" (" + bibPD.year[i] + ")。\n")
        ToutputC += r"\newblock "
        ToutputC += bibPD.title[i]
        # 判断会议是否存在
        if bibPD.organization[i] != "":
            ToutputC += bibPD.organization[i]
        if bibPD.pages[i] != "":
            ToutputC += ("（頁 " + bibPD.pages[i] + "）")
        # 判断是否存在出版社
        if bibPD.location[i] != "" and bibPD.publisher[i] != "":
            ToutputC += ("。" + bibPD.location[i] + "：" + bibPD.publisher[i])
        elif bibPD.location[i] != "":
            ToutputC += ("。" + bibPD.location[i])
        elif bibPD.publisher[i] != "":
            ToutputC += ("。" + bibPD.publisher[i])
        ToutputC += "。\n\n"

    # type ------->>图书类
    elif bibPD.bibType[i] in ["book", "booklet"]:
        ToutputC += (r"\bibitem[" + citeauthor + ", ")
        ToutputC += (bibPD.year[i] + r"]{")
        ToutputC += (bibPD.citelabel[i] + "}\n")
        ToutputC += (refauthor + r"（" + bibPD.year[i] + "）。\n")
        ToutputC += r"\newblock \textbf{ "
        ToutputC += bibPD.title[i] + "}"
        # 判断是否存在出版社
        if bibPD.location[i] != "" and bibPD.publisher[i] != "":
            ToutputC += ("。" + bibPD.location[i] + "：" + bibPD.publisher[i])
        elif bibPD.location[i] != "":
            ToutputC += ("。" + bibPD.location[i])
        elif bibPD.publisher[i] != "":
            ToutputC += ("。" + bibPD.publisher[i])
        ToutputC += "。\n\n"

    # type ------->>书籍类
    elif bibPD.bibType[i] in ["inbook"]:
        ToutputC += (r"\bibitem[" + citeauthor + ", ")
        ToutputC += (bibPD.year[i] + r"]{")
        ToutputC += (bibPD.citelabel[i] + "}\n")
        ToutputC += (refauthor + r"（" + bibPD.year[i] + "）。\n")
        ToutputC += r"\newblock  "
        ToutputC += bibPD.booktitle[i]
        if bibPD.editor[i] != "":
            if " and " in bibPD.editor[i]:
                xx = (bibPD.editor[i].split(" and ")[0].strip() + "等人")
            elif "," in bibPD.editor[i]:
                xx = (bibPD.editor[i].split(",")[0].strip() + "等人")
            else:
                xx = bibPD.editor[i]
            ToutputC += ("載於" + xx + "（主編）")
        ToutputC += (r"，\textbf{" + bibPD.title[i] + r"}")
        if bibPD.pages[i] != "":
            ToutputC += ("（頁 " + bibPD.pages[i] + "）")
        # 判断是否存在出版社
        if bibPD.location[i] != "" and bibPD.publisher[i] != "":
            ToutputC += ("。" + bibPD.editor[i] + ": " + bibPD.publisher[i])
        elif bibPD.location[i] != "":
            ToutputC += ("。" + bibPD.location[i])
        elif bibPD.publisher[i] != "":
            ToutputC += ("。" + bibPD.publisher[i])
        ToutputC += "。\n\n"
    return ToutputC
```
